```python
import pandas as pd
import plotly.graph_objs as go

# Charger les donnes de la variable  reprsenter
#data = pd.read_csv('data.csv')

# Crer une carte gospatiale
fig = go.Figure(go.Choropleth(
    locations = data['Country Code'],
    z = data['Latest water withdrawal data'],
    text = data['Short Name'],
    colorscale = 'YlOrRd',
    autocolorscale=False,
    marker_line_color='white',
    marker_line_width=0.2,
    colorbar_title = 'Latest water withdrawal data'
))

# Ajouter des titres et des tiquettes
fig.update_layout(
    title_text='Carte du monde',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    )
)

# Afficher la carte
fig.show()
```

```python
#Cration d'une colonne quantitative reprsentant le niveau de revenu


modaliteRevenu = data['Income Group'].unique()
print(modaliteRevenu)
mapping = {}
ctp = 1
for ii in modaliteRevenu : 
    mapping[ii] = mapping.get(ii,0)+ctp
    ctp+=1
mapping


data["Income_group_num"] = data['Income Group'].map(mapping)
data["Income_group_num"]
```
